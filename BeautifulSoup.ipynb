BeautifulSoup

Python library used to web scraping, information from the HTML and XML pages.

It helps you:

1) Parse HTML

2) Find tags easily

3) Extract text, links, images, tables, etc.

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.geeksforgeeks.org/python/python-interview-questions/"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

print(soup.title.text)
```

```python
#Find single tag
heading = soup.find("h1")
print(heading.text)
```

```python
#Find multiple tag
para = soup.find_all("p")
for p in para:
  print(p)
```

```python
#Find using class and id

div=soup.find("div", class_="header")
print(div)
```

```python
#Extracting images
imgs = soup.find_all("img")

for img in imgs:
  print(img["src"])
```

```python
#Extracting all link

links = soup.find_all("a")
for link in links:
    print(link.get("href"))
```
